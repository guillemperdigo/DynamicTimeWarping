```python
###ADFuller test
```

```python
# funzione che confronta il p-value con il livello di significatività del test
# e scrive se accettare o rifiutare l'ipotetsi nulla.

def hypothesis(p_value, alpha):
    if p_value<= alpha:
        print("Non accettiamo H0")
    else:
        print("Accettiamo H0")
```

```python
from statsmodels.tsa.stattools import adfuller
adf = adfuller(bitty["High"])
```

```python
adf
```

```python
hypothesis(p_value=round(adf[1],2), alpha=0.05)
```

```python
### KPSS test
```

```python
from statsmodels.tsa.stattools import kpss
kpss_test = kpss(bitty["High"], regression = 'ct', lags = 13, )
```

```python
kpss_test
```

```python
hypothesis(round(kpss_test[1],2), alpha=0.05)
```

```python
### Rendere la serie stazionaria
```

```python
stat = bitty["High"].diff()
```

```python
plt.figure(figsize=(16,8))
plt.plot(stat)
plt.title("Bitty [stazionaria?]")
```

```python
adf_stat = adfuller(stat.values[1:])
print("p-value: ", round(adf_stat[1], 2))
hypothesis(p_value=round(adf_stat[1], 2), alpha=0.05)
```

```python
kpss_stat = kpss(stat.values[1:])
print("p-value: ", round(kpss_stat[1], 2))
hypothesis(p_value=round(kpss_stat[1], 2), alpha=0.05)
```

```python

a = np.arange(1, 100).cumsum()
b = np.arange(5, 104).cumsum()

fig, ax = plt.subplots()
ax.plot(a, color='r')
ax.plot(b, color='b')

ax.set_title("numpy corrcoef test")
plt.show()
```

```python
np.corrcoef([a, b])
```

hm_datapoints = len(hyman.y)
bc_datapoints = len(bitty.Close)

```python
hm_datapoints = len(hyman.y)
bc_datapoints = len(bitty.Close)
```

```python
every = bc_datapoints // hm_datapoints
stretched = np.zeros(bc_datapoints)
for hd in range(hm_datapoints):
    stretched[hd * every] = hyman.y.iloc[hd]

stretched[-1] = hyman.y.iloc[-1]
stretched[np.where(stretched == 0)] = np.nan
```

```python
def nan_helper(y):
    return np.isnan(y), lambda z: z.nonzero()[0]

nans, x = nan_helper(stretched)
stretched[nans] = np.interp(x(nans), x(~nans), stretched[~nans])
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].plot(bitty.Close)
ax[1].plot(stretched)

ax[0].set_title("BitCoin price", fontsize=15)
ax[1].set_title("Hyman Minsky bubble chart(linear interpolation)", fontsize=15)

plt.tight_layout()
plt.show()
```

```python
np.corrcoef([bitty.Close, stretched])
```
